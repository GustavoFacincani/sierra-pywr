```python
import pandas as pd
```

```python
# Stanislaus
filename_in = '../data/Stanislaus River/gauges/storage.csv'
filename_out = '../data/Stanislaus River/gauges/storage_cleaned.csv'
df = pd.read_csv(filename_in, index_col=[0], parse_dates=True).fillna(0)
df.to_csv(filename_out)
df.head()
```

```python
# Tuolumne
filename_in = '../data/Tuolumne River/Observed/Reservoirs/GaugeReservoir_mcm_TUO_R3_old.csv'
filename_out = '../data/Tuolumne River/Observed/Reservoirs/GaugeReservoir_mcm_TUO_R3.csv'
df = pd.read_csv(filename_in, index_col=[0], parse_dates=True).fillna(0)['1950-10-01':'2018-09-30'].replace('#VALUE!', 0)
for c in df.columns:
    df[c] = pd.to_numeric(df[c])
df.to_csv(filename_out)

filename_in = '../data/Tuolumne River/Observed/Streamflow/GaugeStreamFlow_cms_TUO_R3_old.csv'
filename_out = '../data/Tuolumne River/Observed/Streamflow/GaugeStreamFlow_cms_TUO_R3.csv'
df = pd.read_csv(filename_in, index_col=[0], parse_dates=True).fillna(0)['1950-10-01':'2018-09-30'].replace('#VALUE!', 0)
for c in df.columns:
    df[c] = pd.to_numeric(df[c])
df.to_csv(filename_out)
```
