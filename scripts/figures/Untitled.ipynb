```python
%matplotlib inline
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
resultspath = '../../results'
sset = 'Functional Flows test'
basin = 'tuolumne'
climate = 'historical/Livneh'
datapath = os.path.join(resultspath, sset, basin, climate)
filename = 'InstreamFlowRequirement_Flow_mcm.csv'
filepath = os.path.join(datapath, filename)
```

```python
df = pd.read_csv(filepath, header=[0,1], index_col=0, parse_dates=True)
flow = df.xs('IFR at La Grange', level='node', axis=1) / 0.0864 * 35.315
fig, ax = plt.subplots(figsize=(16, 8))
flow.plot(ax=ax)
ax.set_ylabel('Flow (cfs)', fontsize=14)
ax.set_xlabel('Date', fontsize=14)
plt.show()
```
