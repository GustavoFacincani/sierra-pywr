```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
obs = pd.read_csv('./results/observed storage.csv', parse_dates=True, header=0, index_col=0)
sim = pd.read_csv('./results/storage.csv', parse_dates=True, header=0, index_col=0)
fig, axes = plt.subplots(len(sim.columns)-1, 1, figsize=(10, 40))
for i, n in enumerate(obs):
    ax = axes[i]
    ax.plot(obs[n])
    ax.plot(sim[n.replace('/observed storage', '/storage')])
    ax.set_title(n.split('/')[1])
    ax.set_ylabel('Storage (million m3)')
plt.tight_layout()
```
